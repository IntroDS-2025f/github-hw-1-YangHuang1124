````python
# ‰Ω†ÁöÑÁ®ãÂºè

import numpy as np
import matplotlib.pyplot as plt

# Âõ∫ÂÆö‰∫ÇÊï∏Á®ÆÂ≠ê‰ª•‰æøÁµêÊûúÂèØÈáçÁèæ
np.random.seed(768)

N = 200
x = np.random.normal(loc=2.0, scale=1.0, size=N) 
y = x * 0.5 + np.random.normal(scale=0.8, size=N)

size_var = np.abs(np.random.normal(loc=20, scale=10, size=N))
size_scale = size_var ** 2 * 0.5 

scatter = plt.scatter(
    x, 
    y, 
    c=y, 
    s=size_scale, 
    alpha=0.6, 
    cmap='viridis'
)
plt.title("Scatter with Colormap")
colorbar = plt.colorbar(scatter)
colorbar.set_label("y-values")

plt.show()

````
